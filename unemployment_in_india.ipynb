```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime as dt
import calendar
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/Unemployment in India.csv')
df.head()
```

```python
print("Shape of Data: \n{}".format(df.shape))
```

```python
print("Checking for Null Values: \n{}".format(df.isnull().sum()))
```

```python
print("Information about Features: \n{}".format(df.info()))
```

```python
print("Statistical Overview: \n{}".format(df.describe()))
```

```python
df.columns = ["Region","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate (%)","Area"]
df.head()
```

```python
df['Region'].unique()
```

```python
print("Groupby Region: \n{}".format(df.groupby("Region").size()))
```

```python
IND =  df.groupby(["Region"])[['Estimated Unemployment Rate', "Estimated Employed", "Estimated Labour Participation Rate (%)"]].mean()
IND = pd.DataFrame(IND).reset_index()
```

```python
month = IND['Region']
Unemployment_Rate = IND["Estimated Unemployment Rate"]
Labour_Participation_Rate = IND["Estimated Labour Participation Rate (%)"]

fig = go.Figure()

fig.add_trace(go.Bar(x = month, y = Unemployment_Rate, name= "Unemployment Rate"))
fig.add_trace(go.Bar(x = month, y = Labour_Participation_Rate, name= "Labour Participation Rate (%)"))

fig.update_layout(title="Unemployment Rate and Labour Participation Rate",
                  xaxis={"categoryorder":"array", "categoryarray":["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"]})

fig.show()
```

```python
state_stats = df.groupby(['Region'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate (%)']].mean().reset_index()
state_stats = round(state_stats, 2)
state_stats
```

```python
fig = px.box(df,x='Region',y='Estimated Unemployment Rate',color='Region',title='Unemployment rate')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.bar(state_stats, x='Region', y="Estimated Unemployment Rate", color="Region", title="Average Unemploment Rate (State_wise)")
fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.show()
```

```python
region_stats = df.groupby(['Region'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate (%)']].mean().reset_index()
region_stats = round(region_stats, 2)
region_stats
```

```python
before_lockdown = df[(df['Month_int'] >= 1) & (df['Month_int'] <4)]
lockdown = df[(df['Month_int'] >= 4) & (df['Month_int'] <=6)]
af_lockdown = lockdown.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index()

lockdown = before_lockdown.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index()
lockdown['Unemployment Rate after lockdown'] = af_lockdown['Estimated Unemployment Rate']

lockdown.columns = ['Region','Unemployment Rate before lockdown','Unemployment Rate after lockdown']
lockdown
```
